```python
import statistics

# Example data
data = [1, 2, 2, 3, 4, 4, 4, 5, 6, 7]

# Calculate the mean
mean_value = statistics.mean(data)
print(f"Mean: {mean_value}")

# Calculate the median
median_value = statistics.median(data)
print(f"Median: {median_value}")

# Calculate the mode
# Note: statistics.mode() will raise an error if there's no unique mode.
# For multiple modes, use statistics.multimode()
try:
    mode_value = statistics.mode(data)
    print(f"Mode: {mode_value}")
except statistics.StatisticsError:
    multimode_values = statistics.multimode(data)
    print(f"Multimode: {multimode_values}")
```

## Mean, Mode, and Median for CSV Data

Now, let's apply these statistical measures to the `GlobalWeatherRepository.csv` dataset. We'll load the CSV into a pandas DataFrame, identify the numerical columns, and then calculate the mean, median, and mode for each of them.

```python
import pandas as pd
import os

# Construct the full path to the CSV file
csv_file_path = os.path.join(path, 'GlobalWeatherRepository.csv')

# Load the CSV data into a pandas DataFrame
df_weather = pd.read_csv(csv_file_path)

print("First 5 rows of the DataFrame:")
display(df_weather.head())
```

```python
import pandas as pd
import os
csv_file_path = os.path.join(path, 'GlobalWeatherRepository.csv')
df_weather = pd.read_csv(csv_file_path)

print("First 5 rows of the DataFrame:")
display(df_weather.head())
```

```python
import statistics

print("Calculating Mean, Median, and Mode for numerical columns:")
for column in df_weather.select_dtypes(include=['number']).columns:
    print(f"\n--- Column: {column} ---")

    column_data = df_weather[column].dropna()

    if not column_data.empty:
        mean_val = statistics.mean(column_data)
        median_val = statistics.median(column_data)

        try:
            mode_val = statistics.mode(column_data)
            print(f"  Mean: {mean_val:.2f}")
            print(f"  Median: {median_val:.2f}")
            print(f"  Mode: {mode_val:.2f}")
        except statistics.StatisticsError:
            multimode_vals = statistics.multimode(column_data)
            print(f"  Mean: {mean_val:.2f}")
            print(f"  Median: {median_val:.2f}")
            print(f"  Multimode: {multimode_vals}")
    else:
        print("  Column contains no valid numerical data for calculation after dropping NaNs.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a histogram for 'temperature_celsius'
plt.figure(figsize=(10, 6))
sns.histplot(df_weather['temperature_celsius'].dropna(), kde=True, bins=30, color='skyblue')
plt.title('Distribution of Temperature (Celsius)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_weather['temperature_celsius'], y=df_weather['humidity'], alpha=0.6, color='coral')
plt.title('Temperature vs. Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.show()
```
